간단한 자동사냥 RPG 게임을 만들어봅시다! 사용자의 이름을 입력 받아 플레이어를 생성하고, 몬스터들을 차례로 자동사냥하는 게임을 만들어보고자 합니다.

1. 이름, 레벨, 체력, 공격력, 방어력의 속성을 가짐
2. 인스턴스의 현재 체력이 0 이상인지 bool 값을 반환하는 is_alive 메서드 만들기
3. 공격을 받았을 때, (받은 데미지 - 본인의 방어력)만큼 현재 체력이 감소하는 take_damage 메서드 만들기
  - 본인의 방어력이 데미지보다 크다면 체력 감소하지 않음
4. 타겟에게 데미지를 입히는 attack_target 메서드 만들기
5. 데미지는 1부터 공격력 사이의 랜덤한 정수

Q2. Player 클래스와 Monster 클래스 만들기
- Character 클래스를 상속 받는 Player와 Monster 클래스를 만들기

- Player 클래스
  1. Character를 상속 받기
  2. 레벨 1, 체력 100, 공격력 25, 방어력 5로 초기화하기
  3. Player 클래스는 경험치 속성을 추가로 가짐
  4. 인수로 받은 정수 만큼 경험치를 획득하는 gain_experience 메서드 만들기
  5. 현재 경험치가 50이상이면
    - 레벨을 1, 공격력을 10, 방어력을 5씩 올리는 level_up 메서드 만들기

- Monster 클래스
  1. Character를 상속 받기
  2. 몬스터의 레벨에 비례하는 체력, 공격력, 방어력 초기화하기
    - 체력: 10~30 사이의 랜덤한 정수 * 레벨
    - 공격력: 5~20 사이의 랜덤한 정수 * 레벨
    - 방어력: 1~5 사이의 랜덤한 정수 * 레벨

Q3. battle 함수 만들기
- battle 함수 :Player 인스턴스와  Monster 인스턴스를 인수로 받아 둘 중 하나의 체력이 0 이하가 될 때까지 공격을 주고 받는 함수
1. 만약 Player 인스턴스가 살아남았다면
  - Player 인스턴스에 (몬스터 레벨 * 20)만큼의 경험치를 추가
  - player의 레벨업 메서드 호출
  - ‘전투 승리!’를 출력
2. Player 인스턴스가 살아남지 못했을 경우
 ‘전투 패배..’를 출력하기


Q4. main 함수 만들기
- 몬스터의 이름, 레벨이 매핑된 딕셔너리 정의하기
  - monster_dict = {‘슬라임’: 1, ‘고블린’: 2, ‘오크’: 3}
- 사용자로부터 이름을 입력받아 Player 인스턴스 생성하기
- 몬스터 딕셔너리로부터 Monster 인스턴스 생성하기
- 생성된 Monster 인스턴스와 Player 인스턴스가 battle 함수를 통해 전투
  - player는 생성된 몬스터 3마리(슬라임, 고블린, 오크)와 모두 전투해야함
  - 전투 도중에 Player가 사망하면 이후 전투를 진행하지 않고 ‘게임오버’ 출력하고   종료


```python
import random

#Q1. Character 클래스 만들기
class Character:
    def __init__(self, name, level, health, attack, defense): #이름, 레벨, 체력, 공격력, 방어력
        self.name = name
        self.level = level
        self.health = health
        self.attack = attack
        self.defense = defense

    def is_alive(self): #인스턴스의 현재 체력이 0 이상인지 bool 값을 반환하는 is_alive 메서드
        return self.health > 0

    def take_damage(self, damage):
        damage_taken = damage - self.defense  # 실제 받는 데미지 계산
        if damage_taken > 0: # 실제 데미지가 0보다 크면 (방어력보다 받은 데미지가 더 큰 경우)
            self.health -= damage_taken # 체력 감소
        else: # 실제 데미지가 0 이하면 (방어력이 받은 데미지보다 크거나 같은 경우)
           print(f"{self.name}의 방어력이 높아 데미지를 받지 않았다!")

    def attack_target(self, target):  #타겟에게 데미지를 입히는 attack_target 메서드
        damage = random.randint(1, self.attack)
        target.take_damage(damage)
        print(f"{self.name}이(가) {target.name}을 데미지: {damage}만큼 공격했다...! ") #현황 출력
        print(f"남은 체력 - {self.name}: {self.health}, {target.name}: {target.health}")
        if not target.is_alive():
            print(f"{target.name}이(가) 죽었습니다.")

    def show_status(self): #유저와 몬스터의 현황 스테이터스
        print(f"[{self.name}] 레벨: {self.level}, 체력: {self.health}, 공격력: {self.attack}, 방어력: {self.defense}")

#Q2. Player 클래스와 Monster 클래스 만들기
class Player(Character): #Character를 상속 받기
    def __init__(self, name): #레벨 1, 체력 100, 공격력 25, 방어력 5로 초기화하기
        super().__init__(name, 1, 100, 25, 5)
        self.experience = 0 #Player 클래스는 경험치 속성을 추가로 가짐

    def gain_experience(self, amount): #인수로 받은 정수 만큼 경험치를 획득하는 gain_experience 메서드 만들기
        self.experience += amount

    def level_up(self): #현재 경험치가 50이상이면 레벨을 1, 공격력을 10, 방어력을 5씩 올리는 level_up 메서드
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience -= 50
            print(f"{self.name} 레벨업! 현재 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}")

class Monster(Character): #Character를 상속 받기
    def __init__(self, name, level):
        self.name = name
        self.level = level
        #몬스터의 레벨에 비례하는 체력, 공격력, 방어력 초기화하기
        self.health = random.randint(10, 30) * level
        self.attack = random.randint(5, 20) * level
        self.defense = random.randint(1, 5) * level


def battle(Player, Monster):
    print(f"\n{Player.name} vs {Monster.name} 전투 시작!")
    Player.show_status() #전투 시작 전 상태창 출력 (플레이어)
    Monster.show_status() #전투 시작 전 상태창 출력 (몬스터)
    while Player.is_alive() and Monster.is_alive(): #Player 인스턴스와 Monster 인스턴스를 인수로 받아 둘 중 하나의 체력이 0 이하가 될 때까지 공격을 주고 받는 함수
        Player.attack_target(Monster)
        if not Monster.is_alive():
            break
        Monster.attack_target(Player)
    if Player.is_alive():
        Player.gain_experience(Monster.level * 20) #Player 인스턴스에 (몬스터 레벨 * 20)만큼의 경험치를 추가
        Player.level_up()
        print("전투 승리!") #승리시 전투승리
    else:
        print("전투 패배..") #패배시 전투패배

def main():
    monster_dict = {"슬라임": 1, "고블린": 2, "오크": 3} #몬스터의 이름, 레벨이 매핑된 딕셔너리 정의하기
    player_name = input("플레이어 이름을 입력하세요: ") #사용자로부터 이름을 입력받아 Player 인스턴스 생성하기
    player = Player(player_name)
    for monster_name, monster_level in monster_dict.items(): #몬스터 딕셔너리로부터 Monster 인스턴스 생성하기
        if not player.is_alive():
            print("게임오버...") #전투 도중에 Player가 사망하면 이후 전투를 진행하지 않고 ‘게임오버’ 출력하고 종료
            return
        monster = Monster(monster_name, monster_level)
        battle(player, monster)
    print("게임이 종료되었습니다.")

main()
```
